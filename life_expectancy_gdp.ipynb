```python
df.rename(columns={"Life expectancy at birth (years)":"LEABY"},inplace=True)
```

```python
colors = sns.color_palette("hls", n_colors=df.Country.nunique())

fig,axes = plt.subplots(2,3,figsize=(10,6))
plt.subplots_adjust(wspace=0.5, hspace=0.8)
ax1 = plt.subplot(2,3,1)
sns.scatterplot(data=df[df.Country=="Chile"],x="LEABY",y="GDP",color=colors[0])
plt.title("Country : Chile")
plt.ylabel("GDP in trillions of U.S. dollars")
ax2 = plt.subplot(2,3,2)
sns.scatterplot(data=df[df.Country=="China"],x="LEABY",y="GDP",color=colors[1])
plt.title("Country : China")
ax2.set_xticks(range(72,77))
ax3 = plt.subplot(2,3,3)
sns.scatterplot(data=df[df.Country=="Germany"],x="LEABY",y="GDP",color=colors[2])
plt.title("Country : Germany")
ax4 = plt.subplot(2,3,4)
sns.scatterplot(data=df[df.Country=="Mexico"],x="LEABY",y="GDP",color=colors[3])
plt.title("Country : Mexico")
plt.ylabel("GDP in trillions of U.S. dollars")
ax4.set_xticks([75,75.5,76,76.5])
ax5 = plt.subplot(2,3,5)
sns.scatterplot(data=df[df.Country=='United States of America'],x="LEABY",y="GDP",color=colors[4])
plt.title("Country : United States of America")
ax6 = plt.subplot(2,3,6)
sns.scatterplot(data=df[df.Country=="Zimbabwe"],x="LEABY",y="GDP",color=colors[5])
plt.title("Country : Zimbabwe")
fig.legend(labels=df['Country'].unique(), loc='center right', bbox_to_anchor=(1.15,0.5), title="Countries")
plt.show()
plt.clf()
```

```python
fig2,axes2 = plt.subplots(2,3,figsize=(12,6) ,facecolor='#F0F8FF')
plt.subplots_adjust(wspace=0.2, hspace=0.6)

aq1 = plt.subplot(2,3,1)
sns.lineplot(data=df[df.Country == "Chile"],x="Year",y="LEABY",color=colors[0])
aq1.set_xticks(range(2000,2015,2))
plt.xticks(rotation=70)
aq1.set_yticks(range(77,81,1))
aq1.set_facecolor('#FFF8DC')
plt.title("Country : Chile")

aq2 = plt.subplot(2,3,2)
sns.lineplot(data=df[df.Country == "China"],x="Year",y="LEABY",color=colors[1])
aq2.set_xticks(range(2000,2015,2))
plt.xticks(rotation=70)
aq2.set_facecolor('#FFF8DC')
plt.title("Country : China")

aq3 = plt.subplot(2,3,3)
sns.lineplot(data=df[df.Country == "Germany"],x="Year",y="LEABY",color=colors[2])
aq3.set_xticks(range(2000,2015,2))
plt.xticks(rotation=70)
aq3.set_facecolor('#FFF8DC')
plt.title("Country : Germany")

aq4 = plt.subplot(2,3,4)
sns.lineplot(data=df[df.Country == "Mexico"],x="Year",y="LEABY",color=colors[3])
aq4.set_xticks(range(2000,2015,2))
plt.xticks(rotation=70)
aq4.set_facecolor('#FFF8DC')
plt.title("Country : Mexico")

aq5 = plt.subplot(2,3,5)
sns.lineplot(data=df[df.Country == 'United States of America'],x="Year",y="LEABY",color=colors[4])
aq5.set_xticks(range(2000,2015,2))
plt.xticks(rotation=70)
aq5.set_facecolor('#FFF8DC')
plt.title("Country : United States of America")

aq6 = plt.subplot(2,3,6)
sns.lineplot(data=df[df.Country == 'Zimbabwe'],x="Year",y="LEABY",color=colors[5])
aq6.set_xticks(range(2000,2015,2))
plt.xticks(rotation=70)
aq6.set_facecolor('#FFF8DC')
plt.title("Country : Zimbabwe")
plt.show()
plt.clf()
```

```python
fig1,axes1 = plt.subplots(1,1,figsize=(15,6))
sns.barplot(data=df,x="Year",y="LEABY",hue="Country")
plt.xticks(rotation=30)
axes1.legend().set_visible(True)
axes1.legend(loc='center left', bbox_to_anchor=(1, 0.5), title="Countries")
fig1.tight_layout() 
plt.show()
plt.clf()
```
