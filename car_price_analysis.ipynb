```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car_prices.csv")
print(df.head())
```

```python
print(df.make.unique())
```

```python
df.make = df.make.str.capitalize()
```

```python
print(df.make.unique())
```

```python
print(df.dtypes)
```

```python
print(df.info())
```

```python
car_list = ['Bmw',"Audi","Ford","Mercedes"]
```

```python
print(df.year.unique())
```

```python
car_price = df[df.make.isin(car_list)]
```

```python
car_price = car_price[car_price.year >= 2000]
```

```python
car_price.isna().sum()
```

```python
print(car_price.head())
```

```python
fig,axes = plt.subplots(2,2,figsize=(12,6))
plt.suptitle("Car Prices Analysis According To Model", fontsize=16, fontweight='bold')
plt.subplots_adjust(bottom=0.15, left=0.1, right=0.9, top=0.9,hspace=0.5)
ax1 = plt.subplot(2,2,1)
sns.barplot(data=car_price[car_price.make == "Bmw"], x = "year",y="sellingprice", color="skyblue") 
ax1.set_yticks(range(2000,45000,5000))
plt.title("Brand : Bmw")
plt.ylabel("Price U.S. Dollars")
plt.xticks(rotation=30)

ax2 = plt.subplot(2,2,2)
sns.barplot(data=car_price[car_price.make == "Audi"], x = "year",y="sellingprice", color="skyblue") 
ax2.set_yticks(range(2000,45000,5000))
plt.title("Brand : Audi")
plt.ylabel("")
plt.xticks(rotation=30)

ax3 = plt.subplot(2,2,3)
sns.barplot(data=car_price[car_price.make == "Ford"], x = "year",y="sellingprice", color="skyblue") 
ax3.set_yticks(range(2000,33000,5000))
plt.title("Brand : Ford")
plt.ylabel("Price U.S. Dollars")
plt.xticks(rotation=30)

ax4 = plt.subplot(2,2,4)
sns.barplot(data=car_price[car_price.make == "Mercedes"], x = "year",y="sellingprice", color="skyblue") 
ax4.set_yticks(range(2000,40000,5000))
plt.title("Brand : Mercedes")
plt.ylabel("")
plt.xticks(rotation=30)
plt.show()
plt.clf()
```

```python
fig1,axes1 = plt.subplots(1,1,figsize=(12,6))
axs = plt.subplot(1,1,1)
sns.barplot(data = car_price,x="make",y="sellingprice",hue="transmission")
plt.show()
plt.clf()
```
